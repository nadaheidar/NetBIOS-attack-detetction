```python
#imports
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.metrics import f1_score, accuracy_score, classification_report, confusion_matrix, plot_confusion_matrix
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import RandomForestClassifier
from matplotlib import pyplot as plt
```

```python
df= pd.read_csv(path)
```

```python
df_with_no_nulls = df.dropna()
```

```python
dropped_data=df_with_no_nulls
```

```python
dropped_data.drop(dropped_data.columns[[0, 1, 2, 4, 7, 21, 22,85]], axis=1, inplace=True)
```

```python
dropped_data
```

```python
data = dropped_data.sample(frac =.2)
data.head()
```

```python
len(data)
```

```python
x = data.iloc[:,:-1]
y = data.iloc[:,-1]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
import collections
collections.Counter(y_train)
```

```python
from imblearn.over_sampling import SMOTE
sm= SMOTE(random_state=2)
x_train_res, y_train_res= sm.fit_resample(x_train, y_train)
```

```python
import collections
collections.Counter(y_train_res)
```

```python
#model CatBoostClassifier
from catboost import CatBoostClassifier
clf_Cat = CatBoostClassifier()
clf_Cat=clf_Cat.fit(x_train_res, y_train_res)
#clf_Cat.predict_proba(x_test)
clf_Cat_y_pred=clf_Cat.predict(x_test)
```

```python
accuracy_CatBoost = accuracy_score(y_true=y_test, y_pred=clf_Cat_y_pred)*100
print("accuracy_CatBoost : ",accuracy_CatBoost,"%")
```

```python
print(classification_report(y_true=y_test, y_pred=clf_Cat_y_pred))
```

```python
plot_confusion_matrix(clf_Cat, x_test, y_test)
plt.show()
```

```python
import pickle
filename = 'CB.sav'
pickle.dump(clf_Cat, open(filename, 'wb'))

loaded_model = pickle.load(open(filename, 'rb'))
result = loaded_model.score(x_test, y_test)
print(result)
```
